```python
### simpler way of finding relevant information form data
SATScores = [1510, 1510, 1540, 1515, 1500, 1530, 1500, 1540, 1520, 1490, 1520]
def quickly_calc(SATScores):
    df = pd.DataFrame(SATScores)
    mean = df.mean()
    median = df.median()
    sd = df.std(ddof = 0)
    return ("mean:  {}\median : {}\nSD: {}\n" .format(mean[0], median[0], sd[0]))

print(quickly_calc(SATScores))
```
